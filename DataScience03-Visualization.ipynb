```python
# mtcars 데이터셋을 이요한 산점도 그래프 Scatter plot 그리기
import pandas as pd
mtcars = pd.read_csv('./data/mtcars.csv') # 헤더가 있고, 구분자는 컴마
mtcars
```

```python
# print(mtcars.shape)
from IPython.display import display

display(mtcars.head())
display(mtcars.tail())
```

```python
# 기초 통계량 확인하기
mtcars.describe()
```

```python
# mpg(연비) vs wt(중량) 산점도 그리기
plt.scatter(mtcars['wt'],mtcars['mpg'])
plt.xlabel('Weight')
plt.ylabel('Mile per Gallon')
plt.title("Weight vs Mile per Gallon")

plt.show()
```

```python
# 상관 계수
print('상관계수(wt vs mpg)')

mtcars.corr()

mtcars['wt'].corr(mtcars['mpg']) #wt변수와 mpg 변수 사이의 상관 계수
```

```python
# 상자 그림(Box Plot)
plt.boxplot(mtcars['mpg'])

plt.show()

print('mpg 기초 통계량')
display(mtcars['mpg'].describe())
```

```python
# 분위수 확인하기
quantitles = mtcars.quantile([0.25,0.5,0.75])
quantitles
```

```python
#mpg의 1, 3사분위수
first_quartile = mtcars['mpg'].quantile(0.25)
third_quartile = mtcars['mpg'].quantile(0.75)
                        
display(first_quartile)       
display(third_quartile) 

iqr = third_quartile - first_quartile
iqr
```

```python
#mpg의 상단 극단치 경계
#Q3 + IQR*1.5

border_top = third_quartile + (iqr*1.5)

display(border_top)



border_bottom = first_quartile - (iqr*1.5)

display(border_bottom)

```

```python
# mpg가 극단치를 가진 자동차
# mpg가 상단 극단치를 초과한 레코드

outliers = mtcars[mtcars['mpg']>border_top]

outliers
```
