```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup

url = "https://www.dogsofthedow.com/largest-companies-by-market-cap.htm"
header01={ 'User-Agent' : 'Mozilla/5.0'}

# Head 속성에 header01 변수 선성
req = Request(url, headers = header01)
html_content = urlopen(req).read()

soup = BeautifulSoup(html_content,'html.parser')

soup

```

```python
tmp = soup.find_all('td',{'class':'column-2'})
len(tmp)
```

```python
tmp[0].get_text()
```

```python
for i in range(len(tmp)-2):
    print( "시가총액 " + str(i+1) + "위 : " + tmp[i].get_text())
```

```python
coporation_list = [x.get_text() for x in tmp[:-2]]
```

```python
source = soup.find_all('td',{'class':'column-3'})
rank_list = [x.get_text() for x in source[:-2]]
print(rank_list)
```

```python
source = soup.find_all('td',{'class':'column-4'})
market_capital_list = [x.get_text() for x in source[:-2]]
market_capital_list = list(map(lambda x: x.replace(',',''),market_capital_list))
market_capital_list = [float(x) for x in market_capital_list]
print(market_capital_list)
```

```python
source = soup.find_all('td',{'class':'column-5'})
day_increase_list = [x.get_text() for x in source[:-2]]
day_increase_list = [str(x) for x in day_increase_list]
day_increase_list = list(map(lambda x: x.replace('%',''),day_increase_list))
day_increase_list = [float(x) for x in day_increase_list]
print(day_increase_list)
```

```python
source = soup.find_all('td',{'class':'column-6'})
month_increase_list = [x.get_text() for x in source[:-2]]
month_increase_list = [str(x) for x in month_increase_list]
month_increase_list = list(map(lambda x: x.replace('%',''),month_increase_list))
month_increase_list = [float(x) for x in month_increase_list]
print(month_increase_list)
```

```python
source = soup.find_all('td',{'class':'column-7'})
year_increase_list = [x.get_text() for x in source]
year_increase_list = [str(x) for x in year_increase_list]
year_increase_list = list(map(lambda x: x.replace('%',''),year_increase_list))
year_increase_list = [float(x) for x in year_increase_list]
print(year_increase_list)
```

```python
print(len(coporation_list))
print(len(rank_list))
print(len(market_capital_list))
print(len(day_increase_list))
print(len(month_increase_list))
print(len(year_increase_list))
```

```python
import pandas as pd
```

```python
stock_df = pd.DataFrame({'company':coporation_list,
                             'total_stock':market_capital_list,
                             'rank':rank_list,
                             'day_increase':day_increase_list,
                             'month_increase':month_increase_list,
                             'year_increase':year_increase_list})
```

```python
stock_df
```

```python
stock_df.set_index("rank",inplace=True)
stock_df
```

```python
year_stock_df = stock_df.sort_values(by='year_increase', ascending=False)
year_stock_df.head()
```

```python
month_stock_df = stock_df.sort_values(by='month_increase', ascending=False)
month_stock_df.head()
```
