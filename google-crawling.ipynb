```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup

url = "https://www.dogsofthedow.com/largest-companies-by-market-cap.htm"
header01={ 'User-Agent' : 'Mozilla/5.0'}

# Head 속성에 header01 변수 선성
req = Request(url, headers = header01)
html_content = urlopen(req).read()

soup = BeautifulSoup(html_content,'html.parser')

soup

```

```python
tmp = soup.find_all('td',{'class':'column-2'})
len(tmp)
```

```python
tmp[0].get_text()
```

```python
for i in range(len(tmp)-2):
    print( "시가총액 " + str(i+1) + "위 : " + tmp[i].get_text())
```
