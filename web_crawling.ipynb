```python
#멜론 차트 정보 퍼오기
import urllib.request
import urllib.parse
from bs4 import BeautifulSoup

url = "https://www.melon.com/chart/month/index.htm"

header01={ 'User-Agent' : 'Mozilla/5.0'}

# Head 속성에 header01 변수 선성
req = urllib.request.Request(url, headers = header01)
html_content = urllib.request.urlopen(req).read()

soup = BeautifulSoup(html_content,'html.parser')
# print(soup.title.text)

titles = soup.find_all("div", {"class": "ellipsis rank01"}) 
singers = soup.find_all("div", {"class": "ellipsis rank02"}) 
albums = soup.find_all("div",{"class": "ellipsis rank03"})

title_list = []
singer_list = []
album_list = []

rank = 100

for title in titles:
    title_list.append(title.find('a').text)

for singer in singers:
    singer_list.append(singer.find('a').text)

for album in albums:
    album_list.append(album.find('a').text)

for i in range(rank):
    print('%3d위: %s [ %s ] => %s'%(i+1, title_list[i], singer_list[i], album_list[i]))



```
