```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup

url = "https://www.dogsofthedow.com/largest-companies-by-market-cap.htm"
header01={ 'User-Agent' : 'Mozilla/5.0'}

# Head 속성에 header01 변수 선성
req = Request(url, headers = header01)
html_content = urlopen(req).read()

soup = BeautifulSoup(html_content,'html.parser')

soup

```

```python
tmp = soup.find_all('td',{'class':'column-2'})
len(tmp)
```

```python
tmp[0].get_text()
```

```python
for i in range(len(tmp)-2):
    print( "시가총액 " + str(i+1) + "위 : " + tmp[i].get_text())
```

```python
companys = soup.find_all('td',{'class':'column-2'})
company_list = [x.get_text() for x in companys[:-2]]
print(company_list)
```

```python
stocks = soup.find_all('td',{'class':'column-4'})
stocks_list = [x.get_text() for x in stocks[:-2]]
print(stocks_list)
```

```python
stocks_list = list(map(lambda x: x.replace(',',''),stocks_list))
stocks_list = [float(x) for x in stocks_list]
print(stocks_list)
```

```python
month_increase = soup.find_all('td',{'class':'column-6'})
month_increase_list = [x.get_text() for x in month_increase[:-2]]
month_increase_list = list(map(lambda x: x.replace('%',''),month_increase_list))
month_increase_list = [float(x) for x in month_increase_list]
print(month_increase_list)
```

```python
year_increase = soup.find_all('td',{'class':'column-7'})
year_increase_list = [x.get_text() for x in year_increase]
year_increase_list = list(map(lambda x: x.replace('%',''),year_increase_list))
year_increase_list = [float(x) for x in year_increase_list]
print(year_increase_list)
```

```python
import pandas as pd
```

```python

stock_df = pd.DataFrame({"company":company_list,
                         "total_stock":stocks_list,
                         "mont_increase":month_increase_list,
                         "year_increase":year_increase_list})
stock_df
```
