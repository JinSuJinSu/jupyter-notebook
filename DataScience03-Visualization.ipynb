```python
#mapolitlib으로 차트 그리기
#ansccombe의 시각화 강조 주장을 검증
#일반적인 가정: 복수의 데이터셋이 통계 수치, 상관관계, 회귀선이 같다면 비슷한 데이터 셋
#anscombe의 주장: 데이터 시각화 하지 않으면 데이터의 분포를 알기 어렵다 -> 시각화 강조

#데이터 셋 불러오기
import seaborn as sns

#샘플 데이터 셋 목록
print(sns.get_dataset_names())

#anscombe 데이터셋 불러오기
anscombe = sns.load_dataset('anscombe')
anscombe
```

```python
# 데이터를 서브셋의 객수 확인
print(anscombe['dataset'].nunique())
```

```python
# 서브셋 분할하기
ds1 = anscombe[anscombe['dataset']=='I'] #데이터 셋
ds2 = anscombe[anscombe['dataset']=='II']
ds3 = anscombe[anscombe['dataset']=='III']
ds4 = anscombe[anscombe['dataset']=='IV']

ds1.describe()
```

```python
ds2.describe()
```

```python
import matplotlib.pyplot as plt
#데이터셋1 그려보기

# plt.plot(ds1['x'], ds1['y'], 'o')
# plt.plot(ds1['x'], ds1['y'], '*')

plt.plot(ds1['x'], ds1['y'], 'p')
```

```python
#ds2 그래프
plt.plot(ds2['x'], ds2['y'], 'v', color = 'orange')
```

```python
# 그래프의 기본 툴
fig = plt.figure(figsize=(12,12))

fig

# 기본틀에 격자 추가하기
axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)

axes1.plot(ds1['x'],ds1['y'], 'o')
axes2.plot(ds2['x'],ds2['y'], '+', color='orange')
axes3.plot(ds3['x'],ds3['y'], '*', color='green')
axes4.plot(ds4['x'],ds4['y'], '^', color='red')

#서브 틀롯에 타이틀 출력하기
axes1.set_title('Dataset l')
axes2.set_title('Dataset ll')
axes3.set_title('Dataset lll')
axes4.set_title('Dataset lV')

# 기본 틀(Figure)의 타이틀 추가
fig.suptitle("Anscombe Datasets")

# 레이아웃 조정하기
fig.tight_layout()

# 그래프 출력하기
plt.show()
```
