```python
real_data = pd.read_csv('data/최종작업파일/real_data.csv', encoding = 'utf-8')
real_data
```

```python
new_list=[]
for data in real_data['대여소명']:
    if "." in data:        
        index_position = data.index(".")
        new_data=data[index_position+1:]
        new_data=new_data.strip()
        new_list.append(new_data)
    else:
        cnt=0
        for i in range(len(data)):
            if data[i].isdigit():
                cnt+=1
            else:
                break
        new_data=data[cnt+1:]
        new_data=new_data.strip()
        new_list.append(new_data)
    
    
```

```python
new_string = real_data['대여소명'][0]
new_string
```

```python
real_data['대여소명']=new_list
real_data
```

```python
del [new_list]
import gc
gc.collect()
```

```python
final=pd.pivot_table(real_data, index = ['년도','대여소명'], values=['이용건수'],\
                        aggfunc = ['count'])
final
```
