```python
criminal_anal_police = pd.read_csv('data/crime_in_seoul_gu.csv', index_col=0)
criminal_anal_police
```

```python
criminal_anal = pd.pivot_table(criminal_anal_police, index='구별', aggfunc=np.sum)
criminal_anal
```

```python
# 강간 검거율, 강도 검거율, 절도 검거율, 살인 검거율, 폭력 검거율
criminal_anal['강간 검거율'] = (criminal_anal['강간 검거'] / criminal_anal['강간 발생'])*100
criminal_anal['강도 검거율'] = (criminal_anal['강도 검거'] / criminal_anal['강도 발생'])*100
criminal_anal['절도 검거율'] = (criminal_anal['절도 검거'] / criminal_anal['절도 발생'])*100
criminal_anal['살인 검거율'] = (criminal_anal['살인 검거'] / criminal_anal['살인 발생'])*100
criminal_anal['폭력 검거율'] = (criminal_anal['폭력 검거'] / criminal_anal['폭력 발생'])*100

criminal_anal
```

```python
criminal_anal.drop(['강간 검거','강도 검거','살인 검거','절도 검거','폭력 검거'], axis=1, inplace=True)

# del criminal_anal['강간 검거']
# del criminal_anal['강도 검거']
# 위의 방식을 써도 된다.


```

```python
criminal_anal
```

```python
criminal_anal['강간 검거율'] = np.where(criminal_anal['강간 검거율']>100,100,criminal_anal['강간 검거율'])
criminal_anal['강도 검거율'] = np.where(criminal_anal['강도 검거율']>100,100,criminal_anal['강도 검거율'])
criminal_anal['절도 검거율'] = np.where(criminal_anal['절도 검거율']>100,100,criminal_anal['절도 검거율'])
criminal_anal['살인 검거율'] = np.where(criminal_anal['살인 검거율']>100,100,criminal_anal['살인 검거율'])
criminal_anal['폭력 검거율'] = np.where(criminal_anal['폭력 검거율']>100,100,criminal_anal['폭력 검거율'])



criminal_anal
```

```python
for i in range(5,len(criminal_anal.columns)):
    criminal_anal[criminal_anal.columns[i]] = \
    np.where(criminal_anal[criminal_anal.columns[i]]>100,100,criminal_anal[criminal_anal.columns[i]])
    
criminal_anal


# 이것도 좋은 방법

col_list = ['강간 검거율','강도 검거율','절도 검거율','살인 검거율','폭력 검거율']

for column in col_list:
    criminal_anal.loc[criminal_anal[column]>100,column]=100
    
    
# rename
criminal_anal.rename(columns={'강간 발생':'강간','강도 발생':'강도','살인 발생':'살인','절도 발생':'절도','폭력 발생':'폭력'}\
                    , inplace=True)


criminal_anal

# 아래 처럼 해도 된다.
# ncriminal_anal.columns.str.replace('발생','')



    
    

             
 
```
