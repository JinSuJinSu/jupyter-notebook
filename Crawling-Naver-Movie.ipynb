```python
#네이버 상영작 목록 수집
# https://movie.naver.com/movie/running/current.nhn
from urllib.request import urlopen
from bs4 import BeautifulSoup

req = urlopen("https://movie.naver.com/movie/running/current.nhn")
#print(req.getcode())

html = req.read().decode('utf-8')
#print(html)

#Soup 객체 생성
soup = BeautifulSoup(html,'html.parser')

#문서 정보
print(soup.title.text)
```

```python
current_movies = soup.select(".lst_detail_t1>li")
# 클래스 list_detail_t1의 자식들 li
#current_movies[0]

for movie in current_movies:
    #하위로 다시 Select
    titles = movie.select('.lst_dsc > .tit > a')
    
    for title in titles:
        print(title.text, "https://movie.naver.com" + title['href'])
```

```python
#포스터 수집

from urllib.request import urlretrieve
import re

# 포스터 저장 디렉터리 생성
import os

if not os.path.exists("./poster"):
    os.makedirs("./poster")
    
# 포스터 수집

for movie in current_movies:
    #하위로 다시 Select
    img = movie.select('img')[0] # 이미지명 : alt속성 + ".jpg"
    
    #불용 문자제거
    title = re.sub(r'[\/:*?<>{,]','-',img['alt'].strip()) + '.jpg'
    
    # 이미지 다운로드하기
    urlretrieve(img['src'],'./poster/' + title)
    print('.', end='')
else:
    print('download complete!b!!')
    
```
