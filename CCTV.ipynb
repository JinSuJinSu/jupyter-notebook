```python
import pandas as pd
```

```python
CCTV_Seoul = pd.read_csv('data/01. CCTV_in_Seoul.csv', encoding = 'utf-8')
```

```python
CCTV_Seoul.head()
```

```python
CCTV_Seoul.rename(columns={'기관명':'구별'},inplace=True)
CCTV_Seoul.head()
```

```python
pop_Seoul = pd.read_excel('data/01. population_in_Seoul.xls',
                         header=2, usecols='B,D,G,J,N')
pop_Seoul.head()
```

```python
pop_Seoul.columns = ['구별','인구수','한국인','외국인','고령자']
pop_Seoul.head()
```

```python

# 컬럼 이름 : 최근증가율
CCTV_Seoul.sort_values(by='구별', ascending=True, inplace=True)
CCTV_Seoul['최근증가율'] = ((CCTV_Seoul['2014년']+CCTV_Seoul['2015년']+CCTV_Seoul['2016년'])/CCTV_Seoul['2013년도 이전'])*100
CCTV_Seoul.sort_values(by='최근증가율', ascending=False).head(5)
```

```python

# pop_Seoul.drop([0],inplace=True)
pop_Seoul.head(30)
```

```python
pop_Seoul['구별'].unique()
pop_Seoul[pop_Seoul['구별'].isnull()]
pop_Seoul = pop_Seoul.dropna()
pop_Seoul
```

```python
# 외국인 비율
# 고령자 비율
pop_Seoul['외국인 비율'] = (pop_Seoul['외국인']/pop_Seoul['인구수'])*100
pop_Seoul['고령자 비율'] = (pop_Seoul['고령자']/pop_Seoul['인구수'])*100
pop_Seoul.head()
```

```python
pop_Seoul.sort_values(by='인구수', ascending=False).head()
```

```python
pop_Seoul.sort_values(by='외국인 비율', ascending=False).head()
pop_Seoul.rename(columns={'외국인 비율':'외국인비율','고령자 비율':'고령자비율'},inplace=True)
```

```python
pop_Seoul.sort_values(by='고령자비율', ascending=False).head()
```

```python
data_result = pd.merge(CCTV_Seoul,pop_Seoul, on='구별')
data_result.head()
data_result.drop(['2013년도 이전','2014년','2015년','2016년'], axis=1, inplace=True)

# 아래 방식 써도 됨
# del data_result['2014년']
# del data_result['2015년']

data_result
```

```python
import matplotlib.pyplot as plt
```

```python
data_result.sort_values(by='소계', inplace=True)
data_result['소계'].plot(kind='barh',grid=True, figsize=(10,10))
plt.show()
```

```python
## CCTV 비율
data_result.sort_values(by='구별', inplace=True)
data_result['CCTV 비율'] = (data_result['소계']/data_result['인구수'])*100
data_result.head()
```

```python
data_result['CCTV 비율'].sort_values().plot(kind='barh',grid=True, figsize=(10,10))
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['소계'], s=50, color='red')
plt.xlabel('인구수')
plt.xlabel('CCTV 개수')
plt.grid()
plt.show()
```

```python
fp1 = np.polyfit(data_result['인구수'],data_result['소계'],1) #1차 함수 형태

```

```python
fp1
```

```python
f1=np.poly1d(fp1)
fx=np.linspace(100000,700000,100)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data_result['인구수'], data_result['소계'], s=50, color='red')
plt.plot(fx,f1(fx), ls='dashed', lw=3, color='green')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
data_result['오차'] = np.abs(f1(data_result['인구수']) - data_result['소계'])
df_sort=  data_result.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구수'], data_result['소계'], c=data_result['오차'], s=50)
plt.plot(fx,f1(fx), ls='dashed', lw=3, color='green')

for n in range(10):
    plt.text(df_sort['인구수'][n] *1.02, df_sort['소계'][n] *0.98,df_sort.index[n], fontsize=15)
plt.xlabel('인구수')
plt.xlabel('인구당비율')
plt.colorbar()
plt.grid()
plt.show()
```
