```python
import pymysql
from haversine import haversine
```

```python
def merge_two_dicts(x, y):
    z = x.copy()   # start with keys and values of x
    z.update(y)    # modifies z with keys and values of y
    return z

my_dict_final = merge_two_dicts(hotel_dict, landmark_dict)
```

```python
haversine(my_dict_final['남산타워'], my_dict_final['명동성당'], unit='km')
# for i, key in enumerate(my_dict_final.keys()):
#     haversine(my_dict_final[key], my_dict_final['젤리호텔'], unit='km') 
```

```python
# 랜드마크에서 호텔의 거리
for k, v in landmark_dict.items():
    for i, j in hotel_dict.items():
        print('{}에서 {}까지의 거리'.format(k, i))
        print(haversine(v, j, unit='km'))
```

```python
import itertools

# 제일가까운 5개 호텔 찾기
def find_shortest_hotel(landmark):
    
    dis = {}
    
    for i, j in hotel_dict.items(): 
             
        dis[i] = haversine(landmark_dict[landmark], j, unit='km')        

    A = {k: v for k, v in sorted(dis.items(), key=lambda item: item[1])}
    hotel = dict(itertools.islice(A.items(), 5))
    
    return hotel
    
for lm in landmark_list:
    print(find_shortest_hotel(lm))
```

```python
# 주소, 위도, 경도 정보 얻기
tasty_name = []

for name in food_seoul['주소']:
    tasty_name.append(str(name))
```

```python
tasty_name
```

```python
food_address = []
food_lat = []
food_lng = []

for name in tasty_name:
    tmp = gmaps.geocode(name, language='ko')
    food_address.append(tmp[0].get('formatted_address'))
    
    tmp_loc = tmp[0].get('geometry')
    
    food_lat.append(tmp_loc['location']['lat'])
    food_lng.append(tmp_loc['location']['lng'])
```

```python
# 맛집 딕셔너리 
food_dict = {}

for i, name in enumerate(food_seoul['식당명']):
    
    food_dict[name] = (food_lat[i], food_lng[i])

print(food_dict)
```

```python
food_dict.items()
```

```python
food_dict
```

```python
import itertools

# 제일가까운 5개 호텔 찾기
def find_shortest_food(landmark):
    
    dis = {}
    
    for i, j in food_dict.items(): 
             
        dis[i] = haversine(landmark_dict[landmark], j, unit='km')        

    A = {k: v for k, v in sorted(dis.items(), key=lambda item: item[1])}
    food = dict(itertools.islice(A.items(), 10))
    
    return food
    
for lm in landmark_list:
    print(find_shortest_food(lm))
```
