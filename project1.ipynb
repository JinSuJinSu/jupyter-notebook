```python
b_size = 5
train_datagen = ImageDataGenerator(rescale=1./255, horizontal_flip = True, #    50% 
                                  width_shift_range = 0.1, #   10%   
                                  height_shift_range = 0.1,
                                  fill_mode = 'nearest')
```

```python
train_generator = train_datagen.flow_from_directory('C:/image/training/',target_size=(400,600),
                                                    batch_size=b_size, class_mode='binary')
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory('C:/image/test/',target_size=(400,600),
                                                    batch_size=b_size, class_mode='binary')
```

```python
len(test_generator)
```

```python
model = Sequential()
model.add(Conv2D(24, (3, 3), input_shape=(400,600,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(24, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(48, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(48))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(2))
model.add(Activation('softmax'))
```

```python
import cv2, sys
import matplotlib.pyplot as plt
import numpy as np

image_mantis = cv2.imread("C:/opencv_practice/mantis.jpg", cv2.IMREAD_UNCHANGED)
image_gray_mantis = cv2.imread("C:/opencv_practice/mantis.jpg", cv2.IMREAD_GRAYSCALE)

image_mantis = cv2.resize(image_mantis, dsize=(640, 480), interpolation=cv2.INTER_AREA)
image_gray_mantis = cv2.resize(image_gray_mantis, dsize=(640, 480), interpolation=cv2.INTER_AREA)

cv2.imshow('image', image_mantis)
cv2.imshow('image_gray', image_gray_mantis)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
blur_mantis = cv2.GaussianBlur(image_gray_mantis, ksize=(11,11), sigmaX=5)
cv2.imshow('blur_mantis', blur_mantis)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret, thresh1 = cv2.threshold(blur_mantis, 80, 255, cv2.THRESH_BINARY)

cv2.imshow('thresh1', thresh1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
edged_mantis = cv2.Canny(thresh1, 10, 250)
cv2.imshow('edged_mantis', edged_mantis)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
contour_mantis = image_mantis.copy()
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (7,7))
closed_mantis = cv2.morphologyEx(edged_mantis, cv2.MORPH_CLOSE, kernel)
cv2.imshow('closed_mantis', closed_mantis)
cv2.waitKey(0)
```

```python
contours, hierarchy = cv2.findContours(closed_mantis,cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
#   .        
#   

for cnt in contours:
    area = cv2.drawContours(contour_mantis, cnt, -1, (0,255,0), 3)
    
    
cv2.imshow('final_result', area)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
