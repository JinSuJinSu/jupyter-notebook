```python
from urllib.parse import urljoin

rank = []
main_menu = []
cafe_name = []
url_add = []


ranks = soup.find_all("div", {"class": "sammyRank"})

informations = soup.find_all("div", {"class": "sammyListing"})

rank_length = 50

for number in ranks:
    rank.append(number.get_text())
    
# print(rank)

url_base = 'http://www.chicagomag.com'


for information in informations:
    cafe_name.append(re.split(('\n|\r\n'),information.get_text())[1])
    main_menu.append(re.split(('\n|\r\n'),information.get_text())[0])
    
    url_add.append(urljoin(url_base,information.find('a')['href']))
    
# print(url_add)
for i in range(len(ranks)):
    print('%2s위:    cafe : {%s}   menu : {%s}    url : %s'%(rank[i], cafe_name[i], main_menu[i], url_add[i]))



```

```python
len(rank), len(main_menu), len(cafe_name), len(url_add)
```

```python
import pandas as pd
data = {'Rank': rank,
            'Cafe': cafe_name,
            'Menu': main_menu,
            'URL': url_add}

df = pd.DataFrame(data)

df
```

```python
# 작동 시간 알아볼 때

from tqdm.notebook import tqdm

import time

item_total = tqdm(df.index)
price = []
address = []

for n in item_total:
    url = df['URL'][n]
    req = Request(url, headers = header01)
    html_content = urlopen(req).read()

    soup = BeautifulSoup(html_content,'html.parser')
    
    tmp_str = soup.find('p','addy').get_text()
    
    price.append(tmp_str.split()[0][:-1])
    address.append(' '.join(tmp_str.split()[1:-2]))
    
    time.sleep(0.05)
    
```

```python
gmaps.geocode('2109 W. Chicago Ave.,',language='ko')

import numpy as np

for address in df['Address']:   
    if address == 'Multiple':
        lat_list.append(np.nan)
        lng_list.append(np.nan)
        
    else:
        new_address = address + ' ' + 'Chicago'
        address_data = gmaps.geocode(new_address,language='ko')
        lat_list.append(address_data[0]['geometry']['location']['lat'])
        lng_list.append(address_data[0]['geometry']['location']['lng'])

print(lat_list)
print(lng_list)
```

```python

numpy_lat = np.array(lat_list)
numpy_lng = np.array(lng_list)

new_numpy_lat = numpy_lat[~np.isnan(numpy_lat)]
new_numpy_lng = numpy_lng[~np.isnan(numpy_lat)]

# 이 방법으로도 nan값을 제거할 수 있음
# [value for value in x if not math.isnan(value)]

lat_number = np.median(new_numpy_lat)
lng_number = np.median(new_numpy_lng)

# 중앙값 말고 평균값을 집어넣는게 더 효율적일 수 있다.

# lat_number = new_numpy_lat.mean()
# lng_number = new_numpy_lng.mean()

fol_map = folium.Map(location=[lat_number, lng_number], zoom_start=12)

for i in range(len(df.index)):
    if np.isnan(df['lat'][i]) and np.isnan(df['lng'][i]):
        continue
        
    else:
        folium.Marker([df['lat'][i], df['lng'][i]], popup=df['Cafe'][i], 
                      icon=folium.Icon(icon='cloud', color='red')).add_to(fol_map)
fol_map
```
