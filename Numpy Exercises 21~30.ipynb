```python
import numpy as np

def solution(bridge_length, weight, truck_weights):
    answer = 0
    
    # 대기 트럭 리스트
    truck_list = truck_weights
    
    # 도착한 트럭들
    arr_truck_list = []
    
    
    # 운전하고 있는 트럭들
    driving_truck_list = []
    
    
    # 타임 리스트 생성
    # 각 트럭마다 경과 시간을 저장해야 한다.
    time_list = np.array([1])
    
    # 첫 트럭은 어떠한 조건이든 무조건 출발 가능
    driving_truck_list.append(truck_list.pop(0))
    
    # 전체 걸리는 시간 변수 필요
    time=1
    
    while len(truck_list)!=0:
        
        # 시간은 계속 증가
        time+=1
        
        
        if sum(driving_truck_list)+truck_list[0]<=weight:
            driving_truck_list.append(truck_list.pop(0))

            # 모든 트럭들의 경과시간을 1씩 더하고 새로운 트럭을 추가한다.
            time_list+=1
            time_list = np.append(time_list, np.array([1]))
            print(driving_truck_list)

        else:
            # 모든 트럭들의 경과시간을 1씩 더한다.
            time_list+=1

        
        # 건너는 시간과 다리 길이가 같으면 트럭은 운전을 완료한다.        
        if time_list[0]==bridge_length:
            arr_truck_list.append(driving_truck_list.pop(0))
            time_list = np.delete(time_list, 0)
            
        else:
            pass
        
    # 트럭이 전부 도착하지 않았으면 계속 수행해야 함    
    while len(driving_truck_list)!=0:
        # 시간은 계속 증가
        time+=1
        
        time_list+=1
        
        if time_list[0]==bridge_length:
            arr_truck_list.append(driving_truck_list.pop(0))
            time_list = np.delete(time_list, 0)
            
        else:
            pass
        
        
    answer = time
    
    return answer

print(solution(2,10,[7,4,5,6]))
```
